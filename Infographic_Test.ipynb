```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and set background color
fig = plt.figure(figsize=(12, 8), facecolor='#25253c')

# Add title
fig.suptitle('My Infographic', fontsize=24, color='white')

# Create a 2x2 grid for subplots
grid = plt.GridSpec(2, 2, wspace=0.3, hspace=0.3)

# Add a bar chart
ax1 = fig.add_subplot(grid[0, 0])
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 5]
ax1.bar(categories, values, color='skyblue')
ax1.set_title('Bar Chart', color='white')
ax1.set_facecolor('#393d5c')
ax1.tick_params(colors='white')

# Add a pie chart
ax2 = fig.add_subplot(grid[0, 1])
sizes = [30, 20, 25, 25]
ax2.pie(sizes, labels=categories, autopct='%1.1f%%', colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
ax2.set_title('Pie Chart', color='white')

# Add a line graph
ax3 = fig.add_subplot(grid[1, 0])
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax3.plot(x, y, color='lightgreen')
ax3.set_title('Line Graph', color='white')
ax3.set_facecolor('#393d5c')
ax3.tick_params(colors='white')

# Add text information
ax4 = fig.add_subplot(grid[1, 1])
ax4.text(0.5, 0.5, 'Additional Information:\n\nThis is a sample infographic\ncreated using Matplotlib',
         ha='center', va='center', color='white', fontsize=12)
ax4.axis('off')

plt.savefig('infographic.png', facecolor=fig.get_facecolor(), edgecolor='none')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Set the background color
BACKGROUND_COLOR = '#E9967A'  # Light coral color

# Define a function to draw a hexagon
def draw_hexagon(ax, center_x, center_y, radius, text_number, text_label):
    hexagon = patches.RegularPolygon(
        (center_x, center_y), numVertices=6, radius=radius,
        orientation=0, facecolor='white', edgecolor='black', lw=2
    )
    ax.add_patch(hexagon)

    # Add the number and text
    ax.text(center_x, center_y + radius / 2, text_number, ha='center', va='center', fontsize=16, color='black', fontweight='bold')
    ax.text(center_x, center_y - radius / 4, text_label, ha='center', va='center', fontsize=12, color='black', fontweight='bold')

# Function to draw arrows
def draw_arrow(ax, start, end):
    ax.annotate('', xy=end, xytext=start,
                arrowprops=dict(arrowstyle='->', color='white', lw=2))

# Main function to create the infographic
def create_infographic():
    fig, ax = plt.subplots(figsize=(12, 8))

    # Set background color
    fig.patch.set_facecolor(BACKGROUND_COLOR)
    ax.set_facecolor(BACKGROUND_COLOR)

    # Remove axes
    ax.axis('off')

    # Hexagon dimensions
    hexagon_radius = 1.2
    horizontal_spacing = hexagon_radius * 2 * 0.85  # Adjust for overlap
    vertical_spacing = hexagon_radius * 1.7  # Adjust vertical spacing

    # ---Positioning of hexagons---
    start_x = 3
    start_y = 6

    positions = [
        (start_x, start_y, '1', 'LABEL'),
        (start_x + horizontal_spacing, start_y, '2', 'LABEL (VS)'),
        (start_x + 2 * horizontal_spacing, start_y, '3', 'LABEL'),
        (start_x + 2 * horizontal_spacing, start_y - vertical_spacing, '4', 'LABEL'),
        (start_x, start_y - vertical_spacing, '7', 'LABEL')
    ]

    # Draw hexagons
    for x, y, num, label in positions:
        draw_hexagon(ax, x, y, hexagon_radius, num, label)

    # Draw arrows between hexagons
    draw_arrow(ax, (start_x, start_y), (start_x + horizontal_spacing, start_y))
    draw_arrow(ax, (start_x + horizontal_spacing, start_y), (start_x + 2 * horizontal_spacing, start_y))
    draw_arrow(ax, (start_x + 2 * horizontal_spacing, start_y), (start_x + 2 * horizontal_spacing, start_y - vertical_spacing))
    draw_arrow(ax, (start_x + 2 * horizontal_spacing, start_y - vertical_spacing), (start_x, start_y - vertical_spacing))

    # Title - Adjusted position
    ax.text(start_x + 3.5 * horizontal_spacing, start_y, 'MAIN TEXT\ndescription',
            ha='left', va='center', fontsize=24, color='white', fontweight='bold')

    # Adjust the view limits
    ax.set_xlim(0, 15)
    ax.set_ylim(0, 10)

    plt.tight_layout()
    plt.show()

# Run the infographic creator
create_infographic()
```
